# Were covid-19 death rates higher in counties with lower median household income? 

Null Hypothesis:
The rate of covid-19 deaths in a New York county has no relationship with its median household income.

Alternative Hypothesis: The rate of covid-19 deaths in a New York county has a relationship with its median household income.

```python
%%R -i df

#run a linear regression on our data
model <- lm(deaths_by_county_of_residence ~ median_household_income_2018, data = merged_list)
summary(model)
```

# What I hope to learn from the statistical test?

The interaction between the rate of covid deaths on the median household income in New York counties. The dataset reflects covid deaths from April 2020 to 2023 and the median household income in New York's counties in 2018. It also shows a positive correlation that explains that 42% of covid deaths could be explained by the median household income which indicates a rise in covid deaths correlates with a rise in counties with higher median household income. 


The p-value is also very small, this means these results are statistically significant. Also the median-household-income is negatively related to covid death rates in New York state counties.Counties with higher income inequality had higher case and death rates.
